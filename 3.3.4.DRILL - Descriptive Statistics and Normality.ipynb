```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
distributionvar = np.random.binomial(1,.5, 100)

plt.hist(distributionvar)
plt.show()

plt.hist(distributionvar , bins=5, color='c')

# Add a vertical line at the mean.
plt.axvline(distributionvar .mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(distributionvar .mean() + distributionvar .std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(distributionvar .mean()-distributionvar .std(), color='b', linestyle='dashed', linewidth=2) 

# Print the histogram.
plt.show()
```

```python
variable1 = np.random.normal(5,.5, 100)
variable2 = np.random.normal(10,1, 100)
variable3 = variable1+variable2

mean=np.mean(variable3)
sd=np.std(variable3)

plt.hist(variable3)

plt.axvline(x=mean,color='green')
plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')
plt.show()
```
